```python
df1['category'] = df1['category'].apply(map_category_to_english)
df1.category.value_counts()
```

```python
df1.dropna(subset = ['category'], inplace = True)
df1.category.value_counts()
```

```python
# gender: M=male, F=female, U=unisex
df1['gender'] = df1['category'].apply(lambda x: 'F' if 'women' in x.lower() else ('M' if 'men' in x.lower() else 'U'))
df1.gender.value_counts()
```

```python
def determine_gender(row):
    name, description = row["name"].lower(), row["description"].lower()
    has_nam = "nam" in name or "nam" in description
    has_nu = "nữ" in name or "nữ" in description
    if has_nam and not has_nu:
        return "M"
    elif has_nu and not has_nam:
        return "F"
    else:
        return "U"
df1["gender"] = df1.apply(determine_gender, axis=1)
df1.gender.value_counts()
```
